```python
from river.sketch import Histogram
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)

values = np.hstack((
    np.random.normal(-3, 1, 1000),
    np.random.normal(3, 1, 1000),
))

hist = Histogram(max_bins=60)

for x in values:
    hist = hist.update(x)

ax = plt.bar(
     x=[(b.left + b.right) / 2 for b in hist],
     height=[b.count for b in hist],
    width=[(b.right - b.left) / 2 for b in hist]
)
```
